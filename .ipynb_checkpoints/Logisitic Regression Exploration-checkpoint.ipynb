```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

# Predicting Living Situation: Village or City

```python
personality_questions = pd.read_csv('cleaned_youngpeople_data.csv')
personality_questions.head()
```

```python
personality_questions["Education"].unique()
```

```python
cleanup_education = {"Education": {'currently a primary school pupil': 1, 'primary school': 2,
                                  'secondary school': 3, 'college/bachelor degree': 4,
                                  'masters degree': 5, 'doctorate degree': 6}}
```

```python
personality_questions = personality_questions.replace(cleanup_education)
```

```python
living_situation_questions = personality_questions.drop(['Smoking', 'Alcohol', 'Lying', 'Punctuality', 'Only child',
                                                            'Internet usage', 'Gender', 'Left - right handed',
                                                             'House - block of flats'], axis='columns')
```

```python
X_train, X_test, y_train, y_test = train_test_split(living_situation_questions.drop(['Village - town'], axis='columns'),
                                                    living_situation_questions['Village - town'], 
                                                    test_size=0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
y_test.value_counts()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted (0:City, 1:Village)')
plt.ylabel('Truth (0:City, 1:Village)')
plt.title('Living Situation Baseline model- Predicted vs Expected')
#plt.savefig("livingsituationheatmap.png")
plt.show()
```

## Feature Selection

We can plot the feature importance of the columns from our dataset.

```python
importance = model.coef_[0]
plt.bar([x for x in range(len(importance))], importance)
plt.title('Feature Important for Living in a City vs a Village')
plt.show()
```

An even better indicator of what parameters are most important can be sought out using RFECVâ€”recursive feature elimination with cross validation.

```python
from sklearn.feature_selection import RFECV

model = LogisticRegression()
trans = RFECV(model)
logistic_X_trans = trans.fit_transform(X_train, y_train)
columns_retained_RFECV = living_situation_questions.iloc[:, 1:].columns[trans.get_support()].values
```

```python
columns_retained_RFECV
```

```python
RFECV_personality = personality_questions[['Country', 'Village - town']]
RFECV_personality
```

```python
X_train, X_test, y_train, y_test = train_test_split(RFECV_personality.drop(['Village - town'], axis='columns'),
                                                    RFECV_personality['Village - town'], 
                                                    test_size=0.2)
```

```python
RFECVmodel = LogisticRegression()
```

```python
RFECVmodel.fit(X_train, y_train)
```

```python
RFECVmodel.score(X_test, y_test)
```

```python
y_predicted = RFECVmodel.predict(X_test)
cm = confusion_matrix(y_test, y_predicted)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted (0:City, 1:Village)')
plt.ylabel('Truth (0:City, 1:Village)')
plt.title('Final model- Predicted vs Expected')
#plt.savefig("livingsituationheatmap2.png")
plt.show()
```

Our model had a 12% increase in accuracy with feature selection alone. What's most notable in the significant decrease in parameters needed for predicting with 78% accuracy. Only "enjoyment of country music" is needed to predict whether someone lives in a village or city. However, observing the heatmap we can see that the model opts to predict city, giving it a higher accuracy since most survey respondants answered with "city".
